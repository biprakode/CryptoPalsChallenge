```python
import base64

def hex_base64(hex_string : str) -> str:
    bytes_obj = bytes.fromhex(hex_string)
    base64_enc = base64.b64encode(bytes_obj)
    return base64_enc.decode('utf-8')

string = "49276d206b696c6c696e6720796f757220627261696e206c696b65206120706f69736f6e6f7573206d757368726f6f6d"
correct = "SSdtIGtpbGxpbmcgeW91ciBicmFpbiBsaWtlIGEgcG9pc29ub3VzIG11c2hyb29t"

base64_str = hex_base64("49276d206b696c6c696e6720796f757220627261696e206c696b65206120706f69736f6e6f7573206d757368726f6f6d")
print("String: " + string)
print("Base64: " + base64_str)
print("Correct: " + base64_str == correct)
```

```python
def fixed_XOR(hex_str1 : str , hex_str2 : str) -> str:
    bytes1 = bytes.fromhex(hex_str1)
    bytes2 = bytes.fromhex(hex_str2)
    xor = bytes([b1 ^ b2 for b1 , b2 in zip(bytes1, bytes2)])
    return bytes.hex(xor)

str1 = "1c0111001f010100061a024b53535009181c"
str2 = "686974207468652062756c6c277320657965"
correct = "746865206b696420646f6e277420706c6179"
xor = fixed_XOR(str1 , str2)
print("str1: " + str1)
print("str2: " + str2)
print("XOR: " + xor)
print("correct: " + correct == xor)
```

```python
import binascii

def byte_xor(ct , key):
    return bytes([b ^ key for b in ct])

def score_english_text(text):
    freqs = {
        'a': .0817, 'b': .0149, 'c': .0278, 'd': .0425, 'e': .1270, 'f': .0223,
        'g': .0202, 'h': .0609, 'i': .0697, 'j': .0015, 'k': .0077, 'l': .0403,
        'm': .0241, 'n': .0675, 'o': .0751, 'p': .0193, 'q': .0010, 'r': .0599,
        's': .0633, 't': .0906, 'u': .0276, 'v': .0098, 'w': .0236, 'x': .0015,
        'y': .0197, 'z': .0007, ' ': .1300
    }
    return sum([freqs.get(chr(byte).lower(), 0) for byte in text])

def decode_char(ct : str):
    ciphertext = binascii.unhexlify(ct)
    candidates = []
    for key in range(256):
        plaintext = byte_xor(ciphertext, key)
        score = score_english_text(plaintext)
        candidates.append((score, key , plaintext))
    best_score , best_key , best_text = max(candidates, key=lambda x: x[0])
    print("Best text:- ", best_text.decode("utf-8", errors="ignore"))
    print("Best char:- " , chr(best_key))


hex_str = "1b37373331363f78151b7f2b783431333d78397828372d363c78373e783a393b3736"
decode_char(hex_str)
```

```python
def count_repetitions(ciphertext):
    blocks = [ciphertext[i:i+16] for i in range(0, len(ciphertext), 16)]
    total_blocks = len(blocks)
    unique_blocks = len(set(blocks))
    return total_blocks - unique_blocks

file_path = '/run/media/biprarshi/COMMON/files/Crypto/CryptoPals/data/set1q8.txt'
with open(file_path, 'r', ) as f:
    b64_content = f.read().replace('\n', '')
    ciphertext = base64.b64decode(b64_content)
result = count_repetitions(ciphertext)
print("ECB detected in " + str(result))
```
